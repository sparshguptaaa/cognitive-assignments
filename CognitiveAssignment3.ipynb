```python
dataset = {
    'Employee_ID' : [101,102,103,104,105],
    'Name' : ['Alice','Bob','Charlie','Diana','Edward'],
    'Department' : ['HR','IT','IT','Marketing','Sales'],
    'Age' : [29,34,41,28,38],
    'Salary' : [50000,70000,65000,55000,60000],
     'Years_of_Experience' : [4,8,10,3,12],
    'Joining_Date' : ['2020-03-15','2017-07-19','2013-06-01','2021-02-10','2010-11-25'],
    'Gender'  : ['Female','Male','Male','Female','Male'],
    'Bonus'  : [5000,7000,6000,4500,5000],
    'Rating'  : [4.5,4.0,3.8,4.7,3.5]
}
dataframe = pd.DataFrame(dataset)
print(dataframe)

#a) shape of dataframe
print(dataframe.shape)

#b) summary of dataframe
print(dataframe.info())

#c) generate descriptive statistics
print(dataframe.describe())

#d) display first 5 rows and last 3 rows of dataset
print(dataframe.head(5))
print(dataframe.tail(3))

#e)i) average salary of employees
print(dataframe['Salary'].mean())

#e)ii) total bonus paid to all employees
print(dataframe['Bonus'].sum())

#e)iii)youngest employee's age
print(dataframe['Age'].min())

#e)iv) highest performance rating
print(dataframe['Rating'].max())

#f)sort salary column in descending order
print(dataframe.sort_values(by='Salary',ascending=False))

#g)add a new column that categorizes employees based on their performance rating
dataframe['Category'] = ['High' if rating >= 4.5 else 'Good' if rating>=4.0 and rating<4.5 else 'Average' for rating in dataframe['Rating']]
print(dataframe)

#h) missing values in dataframe
print(dataframe.isnull())

#i)rename employee_id to ID
dataframe.rename(columns={'Employee_ID':'ID'},inplace=True)
print(dataframe)

#j)i) finding employees who have experience>5 years
print(dataframe.loc[dataframe['Years_of_Experience']>5])

#j)ii) finding employees who belong to IT department
print(dataframe.loc[dataframe['Department']=='IT'])

#k)add column tax which deducts 10% of salary
dataframe['Tax'] = [taxes for taxes in dataframe['Salary']*0.1]
print(dataframe)

#l) save modified dataframe to new csv file
dataframe.to_csv('modified_dataset.csv',index=False)
```
